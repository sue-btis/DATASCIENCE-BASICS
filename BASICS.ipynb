# **SQL LEER TABLAS**

| **Comando**          | **Definición corta**                                      | **Características**                                                                                          | **Ejemplos**                                                |
|----------------------|-----------------------------------------------------------|---------------------------------------------------------------------------------------------------------------|-------------------------------------------------------------|
| `SELECT`             | Selecciona columnas de una tabla.                        | - Permite especificar las columnas a recuperar.<br>- Soporta filtros y funciones de agregación.              | `SELECT column1, column2 FROM table_name;`<br>`SELECT * FROM table_name;` |
| `WHERE`              | Filtra registros según una condición.                     | - Se usa con `SELECT` para recuperar solo los registros que cumplen con la condición.<br>- Soporta operadores lógicos. | `SELECT * FROM table_name WHERE column1 = 'value';`<br>`SELECT * FROM table_name WHERE column1 > 10;` |
| `ORDER BY`           | Ordena los resultados de la consulta.                     | - Permite ordenar por una o varias columnas.<br>- Puede ordenar en orden ascendente (`ASC`) o descendente (`DESC`). | `SELECT * FROM table_name ORDER BY column1 ASC;`<br>`SELECT * FROM table_name ORDER BY column1 DESC;` |
| `GROUP BY`           | Agrupa los resultados por una o más columnas.             | - Se usa con funciones de agregación como `COUNT`, `SUM`, `AVG`, etc.<br>- Agrupa filas con valores iguales en las columnas especificadas. | `SELECT column1, COUNT(*) FROM table_name GROUP BY column1;`<br>`SELECT column1, AVG(column2) FROM table_name GROUP BY column1;` |
| `HAVING`             | Filtra grupos de registros creados por `GROUP BY`.         | - Se usa para filtrar los resultados después de aplicar `GROUP BY`.<br>- Similar a `WHERE`, pero para grupos. | `SELECT column1, COUNT(*) FROM table_name GROUP BY column1 HAVING COUNT(*) > 5;` |
| `JOIN`               | Combina filas de dos o más tablas basadas en una condición. | - Permite combinar datos relacionados de diferentes tablas.<br>- Existen varios tipos de joins: `INNER JOIN`, `LEFT JOIN`, `RIGHT JOIN`, `FULL JOIN`. | `SELECT * FROM table1 INNER JOIN table2 ON table1.id = table2.id;`<br>`SELECT * FROM table1 LEFT JOIN table2 ON table1.id = table2.id;` |
| `INNER JOIN`         | Devuelve filas cuando hay una coincidencia en ambas tablas. | - Muestra solo las filas con coincidencias en ambas tablas.<br>- Es el tipo de join más común.                | `SELECT * FROM table1 INNER JOIN table2 ON table1.id = table2.id;` |
| `LEFT JOIN`          | Devuelve todas las filas de la tabla izquierda y las filas coincidentes de la tabla derecha. | - Muestra todas las filas de la tabla izquierda y las coincidencias de la tabla derecha.<br>- Las filas sin coincidencias en la tabla derecha tendrán valores `NULL`. | `SELECT * FROM table1 LEFT JOIN table2 ON table1.id = table2.id;` |
| `RIGHT JOIN`         | Devuelve todas las filas de la tabla derecha y las filas coincidentes de la tabla izquierda. | - Muestra todas las filas de la tabla derecha y las coincidencias de la tabla izquierda.<br>- Las filas sin coincidencias en la tabla izquierda tendrán valores `NULL`. | `SELECT * FROM table1 RIGHT JOIN table2 ON table1.id = table2.id;` |
| `FULL JOIN`          | Devuelve todas las filas cuando hay una coincidencia en una de las tablas. | - Muestra todas las filas de ambas tablas, combinadas donde hay coincidencias.<br>- Las filas sin coincidencias en cualquiera de las tablas tendrán valores `NULL`. | `SELECT * FROM table1 FULL JOIN table2 ON table1.id = table2.id;` |

### Ejemplo de uso en SQL:

```sql
-- Seleccionar todas las columnas de una tabla
SELECT * FROM employees;

-- Seleccionar columnas específicas de una tabla
SELECT first_name, last_name FROM employees;

-- Seleccionar columnas con una condición
SELECT * FROM employees WHERE department = 'Sales';

-- Ordenar resultados por una columna
SELECT * FROM employees ORDER BY last_name ASC;

-- Agrupar resultados y contar el número de empleados por departamento
SELECT department, COUNT(*) FROM employees GROUP BY department;

-- Filtrar grupos de resultados
SELECT department, COUNT(*) FROM employees GROUP BY department HAVING COUNT(*) > 5;

-- Combinar datos de dos tablas usando INNER JOIN
SELECT employees.first_name, departments.department_name
FROM employees
INNER JOIN departments ON employees.department_id = departments.department_id;

-- Combinar datos de dos tablas usando LEFT JOIN
SELECT employees.first_name, departments.department_name
FROM employees
LEFT JOIN departments ON employees.department_id = departments.department_id;

-- Combinar datos de dos tablas usando RIGHT JOIN
SELECT employees.first_name, departments.department_name
FROM employees
RIGHT JOIN departments ON employees.department_id = departments.department_id;

-- Combinar datos de dos tablas usando FULL JOIN
SELECT employees.first_name, departments.department_name
FROM employees
FULL JOIN departments ON employees.department_id = departments.department_id;


# **SQL MODIFICAR TABLAS**

| **Comando**         | **Definición corta**                                     | **Características**                                                                                           | **Ejemplos**                                                |
|---------------------|----------------------------------------------------------|----------------------------------------------------------------------------------------------------------------|-------------------------------------------------------------|
| `ALTER TABLE`       | Modifica la estructura de una tabla existente.          | - Permite agregar, eliminar o modificar columnas.<br>- Puede cambiar el nombre de la tabla o las columnas.      | `ALTER TABLE table_name ADD column_name datatype;`<br>`ALTER TABLE table_name DROP COLUMN column_name;`<br>`ALTER TABLE table_name MODIFY COLUMN column_name new_datatype;` |
| `ADD COLUMN`        | Agrega una nueva columna a una tabla.                    | - Se usa con `ALTER TABLE` para añadir nuevas columnas.<br>- Se debe especificar el nombre y el tipo de datos.   | `ALTER TABLE table_name ADD COLUMN new_column datatype;`   |
| `DROP COLUMN`       | Elimina una columna de una tabla.                        | - Se usa con `ALTER TABLE` para eliminar columnas.<br>- La columna y todos sus datos se eliminan permanentemente. | `ALTER TABLE table_name DROP COLUMN column_name;`         |
| `MODIFY COLUMN`     | Cambia el tipo de datos de una columna existente.        | - Se usa con `ALTER TABLE` para modificar el tipo de datos o la definición de una columna.<br>- Puede cambiar restricciones, como tamaño o formato. | `ALTER TABLE table_name MODIFY COLUMN column_name new_datatype;` |
| `RENAME COLUMN`     | Cambia el nombre de una columna en una tabla.            | - Se usa con `ALTER TABLE` para renombrar una columna.<br>- Solo cambia el nombre sin afectar los datos.         | `ALTER TABLE table_name RENAME COLUMN old_column TO new_column;` |
| `RENAME TABLE`      | Cambia el nombre de una tabla.                            | - Se usa para cambiar el nombre de la tabla.<br>- Solo afecta el nombre, no los datos o la estructura.           | `ALTER TABLE old_table_name RENAME TO new_table_name;`    |
| `TRUNCATE TABLE`    | Elimina todos los registros de una tabla.                 | - Rápido para eliminar todos los datos sin eliminar la tabla.<br>- No se puede usar con una cláusula `WHERE`.      | `TRUNCATE TABLE table_name;`                               |
| `DROP TABLE`        | Elimina una tabla y todos sus datos.                      | - Elimina la tabla completamente.<br>- No se puede recuperar la tabla ni sus datos una vez eliminada.             | `DROP TABLE table_name;`                                   |
| `INSERT INTO`       | Agrega nuevas filas a una tabla.                          | - Se usa para insertar uno o más registros en una tabla.<br>- Se deben especificar las columnas y los valores.   | `INSERT INTO table_name (column1, column2) VALUES (value1, value2);` |
| `UPDATE`            | Modifica los registros existentes en una tabla.           | - Se usa para cambiar uno o más valores en registros existentes.<br>- Requiere una cláusula `WHERE` para especificar qué registros se actualizan. | `UPDATE table_name SET column1 = value1 WHERE condition;` |
| `DELETE`            | Elimina registros de una tabla.                           | - Se usa para eliminar uno o más registros.<br>- Requiere una cláusula `WHERE` para especificar qué registros se eliminan. | `DELETE FROM table_name WHERE condition;`                 |

### Ejemplo de uso en SQL:

```sql
-- Agregar una nueva columna
ALTER TABLE employees ADD COLUMN hire_date DATE;

-- Eliminar una columna existente
ALTER TABLE employees DROP COLUMN middle_name;

-- Modificar el tipo de datos de una columna
ALTER TABLE employees MODIFY COLUMN salary DECIMAL(10, 2);

-- Renombrar una columna
ALTER TABLE employees RENAME COLUMN old_name TO new_name;

-- Renombrar una tabla
ALTER TABLE old_employees RENAME TO employees;

-- Eliminar todos los registros de una tabla
TRUNCATE TABLE employees;

-- Eliminar una tabla completamente
DROP TABLE employees;

-- Insertar un nuevo registro en una tabla
INSERT INTO employees (first_name, last_name, department) VALUES ('John', 'Doe', 'Marketing');

-- Actualizar registros existentes en una tabla
UPDATE employees SET department = 'Sales' WHERE department = 'Marketing';

-- Eliminar registros de una tabla
DELETE FROM employees WHERE last_name = 'Doe';


# **FUNCIONES SQL AVANZADAS**

| **Comando / Concepto** | **Definición corta**                                             | **Características**                                                                                          | **Ejemplos**                                                |
|------------------------|------------------------------------------------------------------|----------------------------------------------------------------------------------------------------------------|-------------------------------------------------------------|
| **`DATE` y `TIME`**    | Funciones para manejar y formatear fechas y horas.              | - `DATE` almacena solo la fecha.<br>- `TIME` almacena solo la hora.<br>- Se puede combinar con `DATETIME` para almacenar fecha y hora. | `SELECT CURRENT_DATE;`<br>`SELECT CURRENT_TIME;`<br>`SELECT CURRENT_TIMESTAMP;`<br>`SELECT DATEADD(day, 5, CURRENT_DATE);` |
| **`DATEPART`**         | Extrae una parte específica de una fecha, como el año o el mes. | - Permite obtener partes específicas de fechas.<br>- Usado para realizar cálculos y filtraciones basadas en partes de fechas. | `SELECT DATEPART(year, order_date) FROM orders;`<br>`SELECT DATEPART(month, order_date) FROM orders;` |
| **`SUBQUERY`**         | Consulta dentro de otra consulta.                                | - Permite realizar consultas más complejas.<br>- Puede ser en la cláusula `SELECT`, `WHERE` o `FROM`.<br>- Puede ser `correlated` o `non-correlated`. | `SELECT * FROM employees WHERE department_id IN (SELECT department_id FROM departments WHERE location = 'New York');` |
| **`VIEW`**             | Representación virtual de una tabla basada en una consulta.     | - Se usa para simplificar consultas complejas.<br>- No almacena datos, solo la consulta.<br>- Se puede consultar como una tabla. | `CREATE VIEW employee_sales AS SELECT employee_id, SUM(sales) FROM sales GROUP BY employee_id;`<br>`SELECT * FROM employee_sales;` |
| **`PROCEDURE`**        | Conjunto de instrucciones SQL que se pueden ejecutar como una unidad. | - Permite la reutilización de código.<br>- Puede aceptar parámetros.<br>- Se almacena en la base de datos y se puede ejecutar repetidamente. | `CREATE PROCEDURE GetEmployeeDetails (@EmployeeID INT) AS BEGIN SELECT * FROM employees WHERE employee_id = @EmployeeID; END;`<br>`EXEC GetEmployeeDetails @EmployeeID = 1;` |
| **`ACID TRANSACTIONS`**| Propiedades que aseguran que las transacciones de la base de datos sean confiables. | - **Atomicidad**: Todas las operaciones en una transacción se completan con éxito o ninguna se aplica.<br>- **Consistencia**: La base de datos pasa de un estado válido a otro.<br>- **Aislamiento**: Las transacciones se ejecutan de manera aislada.<br>- **Durabilidad**: Los cambios son permanentes una vez que la transacción se confirma. | N/A (Concepto más que función específica) |
| **`TRANSACTION`**      | Un grupo de operaciones SQL que se ejecutan como una sola unidad. | - Asegura que un conjunto de operaciones se ejecute completamente o no se ejecute en absoluto.<br>- Aumenta la integridad de los datos.<br>- Se usa con comandos `BEGIN`, `COMMIT`, y `ROLLBACK`. | `BEGIN TRANSACTION;`<br>`UPDATE accounts SET balance = balance - 100 WHERE account_id = 1;`<br>`UPDATE accounts SET balance = balance + 100 WHERE account_id = 2;`<br>`COMMIT;`<br>`ROLLBACK;` |

### Ejemplo de uso en SQL:

```sql
-- Obtener la fecha actual
SELECT CURRENT_DATE;

-- Obtener la hora actual
SELECT CURRENT_TIME;

-- Obtener la fecha y hora actuales
SELECT CURRENT_TIMESTAMP;

-- Agregar 5 días a la fecha actual
SELECT DATEADD(day, 5, CURRENT_DATE);

-- Extraer el año de una fecha
SELECT DATEPART(year, order_date) FROM orders;

-- Extraer el mes de una fecha
SELECT DATEPART(month, order_date) FROM orders;

-- Subconsulta en la cláusula WHERE
SELECT * FROM employees WHERE department_id IN (SELECT department_id FROM departments WHERE location = 'New York');

-- Crear una vista
CREATE VIEW employee_sales AS SELECT employee_id, SUM(sales) FROM sales GROUP BY employee_id;

-- Consultar una vista
SELECT * FROM employee_sales;

-- Crear un procedimiento almacenado
CREATE PROCEDURE GetEmployeeDetails (@EmployeeID INT) AS
BEGIN
    SELECT * FROM employees WHERE employee_id = @EmployeeID;
END;

-- Ejecutar un procedimiento almacenado
EXEC GetEmployeeDetails @EmployeeID = 1;

-- Iniciar una transacción
BEGIN TRANSACTION;

-- Actualizar balances de cuentas en una transacción
UPDATE accounts SET balance = balance - 100 WHERE account_id = 1;
UPDATE accounts SET balance = balance + 100 WHERE account_id = 2;

-- Confirmar la transacción
COMMIT;

-- Si ocurre un error, deshacer la transacción
-- ROLLBACK;

## **BINNING**

| **Concepto**      | **Definición corta**                                                                                                                                 | **Características**                                                                                                     | **Ejemplo**                                                                                                          |
|-------------------|-----------------------------------------------------------------------------------------------------------------------------------------------------|-------------------------------------------------------------------------------------------------------------------------|----------------------------------------------------------------------------------------------------------------------|
| **Binning**       | Técnica para agrupar valores continuos en categorías discretas,<br> dividiendo un rango de datos en intervalos o "bins".                                  | - Reducción de la variabilidad: Simplifica los datos al reducir el ruido.<br>- Simplificación de modelos: Facilita el uso en machine learning.<br>- Análisis de distribuciones: Permite ver cómo se distribuyen los datos en intervalos.<br>- Detección de patrones: Ayuda a identificar patrones en los datos. |




````python
import pandas as pd
import numpy as np

data = {'prince': [15, 21, 27, 35, 45, 50, 52, 60]}
df = pd.DataFrame(data)

labels = ['Bajo', 'Medio', 'Alto']  # Etiquetas para los intervalos

bins=np.linspace(min(df["prince"]),max(df["prince"]),4)
df["price-binned"]=pd.cut(df["prince"],bins,labels=labels,include_lowest=True)


print(df)

````
